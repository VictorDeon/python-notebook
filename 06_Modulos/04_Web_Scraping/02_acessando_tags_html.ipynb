## Acessando tags HTML
***

Tags são rotulos usados para informar ao navegador como deve ser apresentado o site.

Todas as tags tem um formato que começa com (<) menor que e termina com (>) maior que

***

```python
from bs4 import BeautifulSoup
```

```python
with open('arquivo.html', 'r') as file:
    soup = BeautifulSoup(file, 'lxml')
```

```python
tag = soup.title
```

```python
print(tag)
```

```python
print(tag.name)
```

```python
tag = soup.p
print(tag.name)
```
